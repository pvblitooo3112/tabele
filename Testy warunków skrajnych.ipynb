| Kategoria ryzyka | Nazwa / Numer Testu | Opis scenariusza / Cel testu |
| :--- | :--- | :--- |
| **Ryzyko kredytowe** | Test RK-1 | Badanie wpływu istotnego wzrostu obliga kredytów zagrożonych na wynik finansowy oraz adekwatność kapitałową. |
| **Ryzyko kredytowe** | Test RK-2 | Badanie wpływu przyrostu rezerw celowych w wyniku przekwalifikowania do sytuacji zagrożonej ekspozycji wobec dwóch największych klientów Banku (zaangażowanie > 10% kapitału Tier I) na wynik finansowy i adekwatność kapitałową. |
| **Ryzyko kredytowe** | Test RK-3 | Badanie wpływu przekwalifikowania 2 największych klientów z najistotniejszej branży do kredytów zagrożonych i utworzenia rezerw ze wskaźnikiem pokrycia min. 50%. |
| **Ryzyko kredytowe** | Test RK-4 | Badanie wpływu istotnego obniżenia się cen nieruchomości na wskaźnik LtV (ocena możliwości zaspokojenia się Banku z najistotniejszego zabezpieczenia). |
| **Ryzyko kredytowe** | Test RK-5 | Badanie wpływu wzrostu rynkowych stóp procentowych na bieżący wskaźnik LtV (wzrost wartości EKZH poprzez wyliczenie i podwyższenie udziału odsetek bieżących o 400 p.b.). |
| **Ryzyko kredytowe** | Test RK-6 | Badanie wpływu wzrostu stóp procentowych na jakość portfela EKZH (założenie: 20% portfela EKZH staje się zagrożone, utworzenie rezerw celowych na poziomie min. 50%). |
| **Ryzyko EKZH** | Scenariusz I (zgodnie z Rekomendacją S) | Wzrost stopy procentowej o poziom wynikający z zaleceń nadzorczych przez min. 12 miesięcy; ocena wpływu na wskaźnik DStI (wyliczenie ekspozycji z DStI > 65%) i jakość portfela EKZH (wyliczenie dodatkowych rezerw celowych i rekalkulacja współczynników). |
| **Ryzyko EKZH** | Scenariusz II (zgodnie z Rekomendacją S) | Wpływ gwałtownej zmiany wartości nieruchomości (obniżenie o x%) na przewidywalną wielkość strat; obliczenie ekspozycji z LtV pow. limitów oraz pow. 100%, z wyliczeniem rezerw celowych dla tych drugich. |
| **Ryzyko DEK** | Scenariusz I (zgodnie z Rekomendacją T) | Wzrost stopy procentowej o poziom z zaleceń nadzorczych; ocena adekwatności maksymalnego wskaźnika DtI i jakości portfela DEK ze zmienną stopą (obliczenie ekspozycji z DtI pow. limitów i wyliczenie różnicy strat oraz rekalkulacja współczynników). |
| **Ryzyko koncentracji** | Test RKK-1 do RKK-5 | Testy wrażliwości na łączny współczynnik kapitałowy dla koncentracji: dużych zaangażowań (RKK-1), branż (RKK-2), DEK (RKK-3), EKZH (RKK-4) oraz otrzymanych zabezpieczeń (RKK-5). |
| **Ryzyko koncentracji** | Test RKK-6 do RKK-10 | Odwrócone testy warunków skrajnych dla koncentracji: dużych zaangażowań (RKK-6), branż (RKK-7), DEK (RKK-8), EKZH (RKK-9) oraz otrzymanych zabezpieczeń (RKK-10). |
| **Ryzyko koncentracji** | T11 (Test scenariuszowy) | Szokowe przekwalifikowanie do należności zagrożonych: 4% dużych zaangażowań, 1% największej branży i 1% największego zabezpieczenia; obliczenie rezerw i skorygowanego współczynnika kapitałowego (minimalny próg to 12,5%). |
| **Ryzyko rynkowe (walutowe)** | Testy spadku kursu PLN | Spadek kursu złotego wobec walut o 30% utrzymujący się przez 12 miesięcy; ocena wpływu na pozycję całkowitą Banku oraz na wynik z pozycji wymiany. |
| **Ryzyko rynkowe (walutowe)** | Odwrotne testy warunków skrajnych | Określenie maksymalnej zmiany kursu walutowego, który nie powoduje przekroczenia całkowitej pozycji walutowej tworzącej wymóg kapitałowy. |
| **Ryzyko operacyjne** | Test RO-I (T1) | Utrata placówek banku i zdolności operacyjnych na skutek katastrofy naturalnej lub ataku terrorystycznego. |
| **Ryzyko operacyjne** | Test RO-2 (T2) | Zatrzymanie operacji w banku internetowym na 2 doby skutkujące utratą zdolności do wykonywania operacji. |
| **Ryzyko operacyjne** | Test RO-3 (T3) | Analizy scenariuszowe negatywnych zdarzeń (np. atak DDoS, Ransomeware, masowe nieobecności przez epidemię, nieobecność IT podczas awarii). |
| **Ryzyko operacyjne** | Test RO-4 (T4) | Test odwrotny wskazujący zdarzenia najbardziej krytyczne dla Banku przekraczające jego odporność na ryzyko. |
| **Ryzyko stopy procentowej** | Zmiana stóp o 250 p.b. | Zmiana wyniku z tytułu odsetek w skali 12 miesięcy przez skok stóp o 250 punktów bazowych oraz niedopasowanie między stopami (ryzyko bazowe). |
| **Ryzyko stopy procentowej** | Odwrócony test stóp referencyjnych | Obliczanie maksymalnej zmiany stóp, przy której Bank nie ponosi straty brutto (zmiana wyniku odsetkowego = planowany zysk brutto). |
| **Ryzyko stopy procentowej** | Interakcja z ryzykiem kredytowym | Zwiększenie kredytów zagrożonych o 5%, co pomniejsza proporcjonalnie aktywa oprocentowane powyżej 2,5%. |
| **Ryzyko stopy procentowej** | Interakcja z ryzykiem płynności | Wzrost zobowiązań pozabilansowych, odpływ depozytów niskooprocentowanych i ich zastąpienie tymi z wyższym oprocentowaniem. Szacowany wpływ spadku stóp o 250 p.b. |
| **Ryzyko stopy procentowej** | Szoki EVE (wartość ekonomiczna kapitału) | Badanie wpływu 6 scenariuszy szokowych (np. równoległy wzrost/spadek stóp, stromy wzrost/spadek krzywej) na EVE oraz dodatkowy test dla zmiany 400 p.b. |
| **Ryzyko stopy procentowej** | Test odwrócony (EVE) | Obliczanie max. zmiany stóp dla której zachowany jest limit spadku wartości ekonomicznej kapitału = 15% Kapitału Tier I. |
| **Ryzyko płynności i finansowania** | Test LCR i Test odwrotny LCR | Testowanie przestrzegania miar LCR (wymóg pokrycia wypływów netto); możliwość wycofania środków kanałami elektronicznymi oraz odwrotny test przestrzegania limitów. |
| **Ryzyko płynności i finansowania** | Test NSFR i Test odwrotny NSFR | Testowanie wskaźnika stabilnego finansowania netto (wskaźnik płynności długoterminowej) oraz odwrotny test jego przestrzegania. |
| **Ryzyko płynności i finansowania** | Różne dynamiki i pogorszenia portfela | Serie testów warunków skrajnych zakładających m.in.: spadek depozytów ogółem o 20%, pogorszenie portfela kredytowego, spadek depozytów terminowych i negocjowanych. |
| **Ryzyko płynności i finansowania** | Test prognoz planu finansowego | Kształtowanie wskaźników płynności w oparciu o plan finansowy Banku na przyszły rok. |
| **Ryzyko nadmiernej dźwigni** | Test straty finansowej | Ocena wpływu straty na wskaźnik dźwigni na skutek zmiany funduszy własnych. |
| **Ryzyko biznesowe** | Niewykupienie instrumentów | Analiza wpływu niewykupienia instrumentów finansowych na wynik finansowy. |
| **Ryzyko biznesowe** | Test odwrotny (Inwestycje) | Badanie poziomu spadku inwestycji (wynik zerowy) i spadek notowań wymagany do spadku TCR poniżej 12,50%. |
| **Ryzyko biznesowe** | Scenariusz RBIZ 1 do RBIZ 3 | RBIZ 1: zmiany w działalności handlowej; RBIZ 2: spadek wyniku odsetkowego i wzrost kosztów rezerw; RBIZ 3 (odwrotny): spadek wolumenów prowadzący do zerowego wyniku lub spadku TCR. |
| **Ryzyko kapitałowe** | Test szokowego obniżenia kapitału | Obniżenie funduszy własnych o 2% na skutek zdarzeń zewnętrznych i wewnętrznych; ocena wpływu na łączny współczynnik. |
| **Ryzyko kapitałowe** | Wpływ przekwalifikowania ekspozycji detalicznych | Wpływ przekwalifikowania 10% ekspozycji detalicznych do zagrożonych na łączny współczynnik kapitałowy (zmiana wagi ryzyka do 100%). |
| **Ryzyko kapitałowe** | Wzrost wymogów kapitałowych | Test wzrostu kosztów ryzyka operacyjnego i dodatkowych wymogów do poziomu przewyższającego minimalne wymogi o 2%. |
| **Ryzyko kapitałowe** | Test odwrotny | Badający możliwości pokrycia straty Banku posiadanymi funduszami własnymi (obrazuje odporność na wzrost ryzyka). |
